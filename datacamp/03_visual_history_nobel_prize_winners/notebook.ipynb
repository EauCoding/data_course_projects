## Question 2: Which decade had the highest ratio of US-bord Nobel Prize winners to total winners in all categories?

```python
# Add decade column
data["decade"] = (data.year // 10) * 10
data.head()
```

```python
# Add US-born column
data["us_born"] = np.where(data.birth_country == "United States of America", "US-born", "Non-US-born")
data.head()
```

```python
# Count through decades and birth country
agg_data = data.groupby(["decade", "us_born"])["category"].count().reset_index(name='category_count')
agg_data = agg_data.pivot(index="decade", columns="us_born", values="category_count").reset_index()
agg_data["us_born_ratio"] = agg_data["US-born"] / (agg_data["Non-US-born"] + agg_data["US-born"])
max_decade_usa = agg_data.sort_values("us_born_ratio", ascending=False).iloc[0, 0]
max_decade_usa
```

## Question 3: Which decade and Nobel Prize category combination had the highest proportion of female laureates?

```python
# Add a female laureate count column
data["female_laureate"] = np.where(data["sex"] == "Female", 1, 0)
data.head()
```

```python
# Count through decade and female laureates
agg_data = data.groupby(["decade", "category"])["female_laureate"].mean().reset_index(name="female_laureate_count").sort_values("female_laureate_count", ascending=False)
max_female_dict = {agg_data.iloc[0, 0]: agg_data.iloc[0, 1]}
max_female_dict
```

## Question 4: Who was the first woman to receive a Nobel Prize, and in what category?

```python
# Get the name of the first woman and the category of the prize
result = data[data["sex"] == "Female"].sort_values("year").head(1)
first_woman_name = result.iloc[0, 7]
first_woman_category = result.iloc[0, 1]

print(first_woman_name, first_woman_category)
```
